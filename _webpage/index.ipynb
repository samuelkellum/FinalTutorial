```python
# Load Numpy
import numpy as np
# Load MatPlotLib
import matplotlib
import matplotlib.pyplot as plt
# Load Pandas
import pandas as pd
```

The source of the csv file: https://data.nola.gov/Public-Safety-and-Preparedness/NOPD-Misconduct-Complaints/gz2m-ef5u

```python
police_complaint_df = pd.read_csv('../_data/NOPD_Misconduct_Complaints.csv')
police_complaint_df.head()
```

```python
#Tidying data by dropping columns we don't plan on using
del police_complaint_df['Incident Type']
del police_complaint_df['Complaint Tracking Number']
del police_complaint_df['Date Complaint Received by NOPD (PIB)']
del police_complaint_df['Date Complaint Occurred']
del police_complaint_df['Date Complaint Investigation Complete']
del police_complaint_df['Bureau of Complainant']
del police_complaint_df['Shift of Complainant']
del police_complaint_df['Unique Officer Allegation ID']

police_complaint_df.head()
```

```python
#Since this data frame does not contain useful information (only four NaN values), we will drop this column as well
police_complaint_df['Officer years of service'].describe()
```

```python
del police_complaint_df['Officer years of service']
```

```python
police_complaint_df["Investigation status"].value_counts().plot.pie()
```

```python
police_complaint_df["Complainant Ethnicity"] = police_complaint_df["Complainant Ethnicity"].map({
    'Black': 'Black',
    'White': 'White',
    'Asian': 'White',
    'Race-Unknown': np.nan,
    'Hispanic': 'Hispanic',
    'b': 'Black',
    'white': 'White',
    'Indian': 'Indian',
    'w': 'White',
    'W': "White",
    'black': 'Black',
    'Unknown': np.nan,
    'Unkown': np.nan,
    np.nan: np.nan
})
police_complaint_df["Complainant Ethnicity"].value_counts().plot.pie()
```
